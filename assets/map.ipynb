```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# Load GeoJSON
world = gpd.read_file("world-fixed.geojson")

# Load CSV (make sure the country names match!)
tariffs = pd.read_csv("tariffs_data.csv")

# Merge using common name key
merged = world.merge(tariffs, how="left", left_on="name", right_on="Country")

# Plot
fig, ax = plt.subplots(figsize=(20, 10))
merged.plot(column="total_mapping", cmap="Reds", linewidth=0.8, ax=ax, edgecolor="0.8", legend=True)
plt.title("Trump Tariffs by Country", fontsize=18)
plt.axis("off")
plt.show()
```
